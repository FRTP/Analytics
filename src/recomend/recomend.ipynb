```python
class User:
    def __init__(self, _user_id):
        self.__user_id = _user_id
        self.__result = 0
        self.__list_of_bids = []
        self.__list_of_diff_result = []
        self.__index = 0
        
    def gUserId(self):
        return self.__user_id
    def sUserId(self, _user_id):
        self.__user_id = _user_id
        
    def gUserIndex(self):
        return self.__index
    def sUserIndex(self, _user_index):
        self.__index = _user_index

    def gResult(self):
        return self.__result
    def sResult(self, _result):
        self.__result = _result
        
    def gListOfDiffResult(self):
        return self.__list_of_diff_result
        
    def updateResult(self, _deltaResult):
        self.__result += _deltaResult
    def updateListDiffResult(self, _deltaResult):
        self.__list_of_diff_result.append(_deltaResult)
        
    def appendListOfBids(self, _bid):
        self.__list_of_bids.append(_bid)
    def gListOfBids(self):
        return self.__list_of_bids
```

```python
class UsersDataBase:
    def __init__(self):
        self.dict_of_users = dict()
    def update(self, _data_clear):
        if type(_data_clear) != list:
            raise Exception('Too dirty')
        for i in range(len(_data_clear)):
            self.__updateKey(_data_clear[i])
                
    def __updateKey(self, _user_request):
        _id = _user_request['user_id']
        diffResult = (100 * (_user_request['current_price'] - _user_request['start_price']) /
        _user_request['start_price'] *
        np.sign(_user_request['target_price'] - _user_request['start_price']))
        if _id not in self.dict_of_users:
            user = User(_id)
            self.dict_of_users[_id] = user
        _user = self.dict_of_users[_id]
        if len(_user.gListOfDiffResult()) > 10:
            _user_index = (100 * np.mean(_user.gListOfDiffResult()) / np.std(_user.gListOfDiffResult()) /
                           np.sqrt(len(_user.gListOfDiffResult())))
            _user.sUserIndex(_user_index)
        _user.appendListOfBids(_user_request)
        _user.updateResult(diffResult)
        _user.updateListDiffResult(diffResult)
        
```

```python
usersDataBase =  UsersDataBase()
```

```python
usersDataBase.update(data_clear)
```
